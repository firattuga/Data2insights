```python
## HOW to plot Distrubition---->Logic is here!! FIROOOO
import numpy as np
import matplotlib.pyplot as plt
mu = 3
std = 1
#distrubition numbers and fixing them with seed
np.random.seed(1)
dist_numbers = np.random.normal(mu,std,1000)


#now one has to create x axis space for theoretical plot. and for each point there has to be a number corresponding to pdf function!!
x = np.linspace(mu - 4*std , mu + 4*std , 1000)
pdf = (np.sqrt(2*np.pi*std**2))**-1 * np.exp(-0.5* ((x-mu)/std)**2)

##NOWW PLOT this shit!

plt.figure(figsize=(10, 6))
#theoretical plot
plt.plot(x,pdf,color = 'red', label = f'Gaussian distribution with  = {mu} and  = {std}',linewidth=2)
#plot barrr
plt.hist(dist_numbers,density = True,bins = 12,color = 'skyblue',alpha = 0.7,label = "Distrubition of numerical choices" )
plt.xlabel("X values")
plt.ylabel("PDF values")
plt.title('Gaussian distrubition with numerical proof')
plt.grid(alpha = 0.2)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Data Generation (Re-using previous setup) ---
np.random.seed(42) 
mu_actual = 3
std_actual = 1
dist_numbers = np.random.normal(mu_actual, std_actual, 100)

# --- Theoretical PDF Curve ---
x = np.linspace(mu_actual - 4 * std_actual, mu_actual + 4 * std_actual, 1000)
pdf = (1 / (std_actual * np.sqrt(2 * np.pi))) * \
      np.exp(-0.5 * ((x - mu_actual) / std_actual)**2)

# --- Subplot Creation and Plotting ---

# Create a figure with 2 rows and 1 column. 
# 'axes' will be an array containing the two Axes objects: [ax1, ax2]
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 8))

# Define the individual plots
ax1 = axes[0] # The top subplot (for the Histogram)
ax2 = axes[1] # The bottom subplot (for the PDF Curve)

# --- Plot 1: Histogram (Top Plot) ---
ax1.hist(dist_numbers, bins=12, density=True, alpha=0.7, color='skyblue',
         label='Sample Histogram (100 points)')
ax1.set_title('Distribution of Sampled Data (Histogram)')
ax1.set_ylabel('Probability Density')
ax1.grid(True, alpha=0.2)
ax1.legend()

# --- Plot 2: Theoretical PDF Curve (Bottom Plot) ---
ax2.plot(x, pdf, color='red', linewidth=2,
         label=f'Theoretical Gaussian PDF (={mu_actual}, ={std_actual})')
ax2.set_title('Theoretical Normal Probability Density Function',)
ax2.set_xlabel('Value')
ax2.set_ylabel('Probability Density')
ax2.grid(True, alpha=0.2)
ax2.legend(loc = 'upper left')

# Adjust layout to prevent titles/labels from overlapping
plt.tight_layout() 

plt.show()
```
